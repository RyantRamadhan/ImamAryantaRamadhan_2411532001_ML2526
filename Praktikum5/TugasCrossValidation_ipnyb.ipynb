```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, cross_validate
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
```

```python
train_url = 'https://raw.githubusercontent.com/RyantRamadhan/ImamAryantaRamadhan_2411532001_ML2526/refs/heads/main/Praktikum5/train.csv'
train = pd.read_csv(train_url)

print('Train shape:', train.shape)
train.head()
```

```python
df_clean = train.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])

from sklearn.impute import SimpleImputer

imputer_age = SimpleImputer(strategy='median')
df_clean['Age'] = imputer_age.fit_transform(df_clean[['Age']])

imputer_embarked = SimpleImputer(strategy='most_frequent')
df_clean['Embarked'] = imputer_embarked.fit_transform(df_clean[['Embarked']]).ravel()

imputer_fare = SimpleImputer(strategy='median')
df_clean['Fare'] = imputer_fare.fit_transform(df_clean[['Fare']])

df_clean['Sex'] = df_clean['Sex'].map({'male': 0, 'female': 1})

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_clean['Embarked'] = le.fit_transform(df_clean['Embarked'])

print(df_clean.info())
```

```python
X = df_clean.drop('Survived', axis=1)
y = df_clean['Survived']
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

scoring = {
    'accuracy' : make_scorer(accuracy_score),
    'precision' : make_scorer(precision_score),
    'recall' : make_scorer(recall_score),
    'f1_score' : make_scorer(f1_score)
}

results = cross_validate(model, X, y, cv=kfold, scoring=scoring)
```

```python
print(f"Akurasi pada masing-masing fold : {np.round(results['test_accuracy'], 4)}")
print(f"Rata-rata Akurasi               : {results['test_accuracy'].mean():.4f}")
print(f"Rata-rata Presisi (Precision)   : {results['test_precision'].mean():.4f}")
print(f"Rata-rata Recall                : {results['test_recall'].mean():.4f}")
print(f"Rata-rata F1-Score              : {results['test_f1_score'].mean():.4f}")
```
